If your target values are large, this MSE might be acceptable.

2. **R-squared (R²) score:**  
   - **Value: 0.9511**
   - **Interpretation:** R² indicates the proportion of variance in the dependent variable that the model explains. A value of 0.9511 means that the model explains approximately 95.11% of the variance in the data. This is a very strong value, implying that the model fits the data well and is able to capture most of the variance in the dependent variable.

3. **Slope:**  
   - **Value: 9337.38**
   - **Interpretation:** The slope represents the change in the dependent variable for each unit change in the independent variable. In this case, for every 1 unit increase in the independent variable, the predicted dependent variable increases by approximately 9337.38. This indicates a strong positive relationship between the predictor and the outcome.

4. **Intercept:**  
   - **Value: 25,755.10**
   - **Interpretation:** The intercept is the value of the dependent variable when the independent variable is zero. In your model, when the independent variable is 0, the predicted value of the dependent variable is around 25,755.10.

### Overall Summary:
- The model fits the data well, as indicated by the high R² score (0.9511).
- The large MSE suggests that there might be some large residuals or that the target variable itself has a large scale.

1. **Mean Squared Error (MSE):**  
   - **Value:** `4919031.07`
   - **Interpretation:** This is the average of the squared differences between the actual values (`y_test`) and the predicted values (`pred`). A smaller MSE indicates that the predictions are closer to the actual data points. In your case, the MSE of around 4.92 million suggests that on average, the squared prediction errors are large, meaning the model has some room for improvement.

2. **R-squared (R²) Score:**  
   - **Value:** `0.757`
   - **Interpretation:** This value measures how well the regression model explains the variability of the target variable. An R² of 0.757 means that 75.7% of the variation in the dependent variable (`y_test`) is explained by the model. This is a decent score, indicating the model is fairly good at explaining the variability, but there is still about 24.3% of the variability unaccounted for.

3. **Slope (Coefficient):**  
   - **Value:** `[1317.31]`
   - **Interpretation:** This is the value of the regression coefficient (or slope) for the predictor variable in the model. For every unit increase in the independent variable, the dependent variable increases by approximately 1317.31 units. This shows the strength and direction of the relationship between the independent variable and the dependent variable.

4. **Intercept:**  
   - **Value:** `-13632.69`
   - **Interpretation:** This is the predicted value of the dependent variable when the independent variable is zero. In this case, when the independent variable is 0, the dependent variable is predicted to be around -13632.69. This is essentially the starting point of the linear model on the y-axis.

### Summary:
- The model explains about 75.7% of the variance in the target variable.
- For each increase of one unit in the independent variable, the dependent variable increases by around 1317.31 units.
- When the independent variable is zero, the model predicts the dependent variable to be -13632.69.
- The relatively high MSE indicates that the model’s predictions are off by a significant amount on average.

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x=np.array([[i for i in range(500 ,1500)]]).reshape(1000,1)
y=np.array([[i*3 +2 for i in range(500,1500)]]).reshape(1000,1)
```

```python
lr = LinearRegression()

# Generating feature (x) and target (y) arrays
x = np.array([[i for i in range(500, 1500)]])
y = np.array([[i * 3 + 2 for i in range(500, 1500)]]).reshape(1000, 1)

# Reshaping x for fitting the model
x1 = x.reshape(1000, 1)

# Fitting the model
lr.fit(x1, y)

# Predicting using the model
preds = lr.predict(x1)

# Calculating Mean Squared Error
mse = mean_squared_error(preds, y)
print("Mean Squared Error:", mse)
r2=r2_score(preds,y)
r2
```
